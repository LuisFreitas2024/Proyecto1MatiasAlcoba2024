```python
import re
import os

# Ruta de la carpeta que contiene los archivos HTML
carpeta_datos = '../data/'

# Lista de nombres de archivos HTML
archivos = [
    'baby1990.html',
    'baby1992.html',
    'baby1994.html',
    'baby1996.html',
    'baby1998.html',
    'baby2000.html',
    'baby2002.html',
    'baby2004.html',
    'baby2006.html',
    'baby2008.html'
]

# Inicializar una lista para contener los datos de fecha de nacimiento, nombre y ranking
lista_datos = []

# Iniciar el documento Markdown
markdown_result = "# Nombres Populares por AÃ±o\n\n"

# Iterar sobre los archivos
for archivo_html in archivos:
    ruta_completa = os.path.abspath(os.path.join(carpeta_datos, archivo_html))
    
    with open(ruta_completa, 'r', encoding='utf-8') as archivo:
        contenido = archivo.read()

    # Encontrar la fecha de nacimiento con una expresiÃ³n regular
    fecha_nacimiento_match = re.search(r'Popularity in (\d{4})</h3>', contenido)
    if fecha_nacimiento_match:
        fecha_nacimiento = fecha_nacimiento_match.group(1)

        # Encontrar los nombres y sus rangos con expresiones regulares
        nombres_match = re.findall(r'<td>(\d+)</td><td>(\w+)</td><td>(\w+)</td>', contenido)

        # Agregar los datos a la lista
        for ranking, nombre_nino, nombre_nina in nombres_match:
            lista_datos.append((fecha_nacimiento, nombre_nino, ranking))
            lista_datos.append((fecha_nacimiento, nombre_nina, ranking))

# Ordenar la lista alfabÃ©ticamente por nombre
lista_datos.sort(key=lambda x: x[1])

# Agregar los resultados como una tabla Markdown
markdown_result += "| Fecha de Nacimiento | Nombre | Ranking |\n"
markdown_result += "|--------------------|--------|---------|\n"

for dato in lista_datos:
    markdown_result += f"| {dato[0]} | {dato[1]} | {dato[2]} |\n"

# Imprimir el resultado en formato Markdown en el cuaderno
from IPython.display import Markdown
Markdown(markdown_result)
```
